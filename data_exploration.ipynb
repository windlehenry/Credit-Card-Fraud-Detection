# Exploratory Data Analysis (EDA)

This notebook explores the credit card transaction dataset to understand:
- The structure and shape of the data
- Class imbalance (fraud vs. non-fraud)
- Feature distributions and possible patterns
- Whether any preprocessing or transformation is required

We will not clean or transform the data here — this is strictly observation.

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter Notebook settings
%matplotlib inline
sns.set(style="whitegrid")
```

## 1. Dataset Overview

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

This dataset contains transactions made by credit card holders in Europe. It includes 284,807 transactions and 31 columns:
- `Time`: seconds since the first transaction
- `Amount`: transaction value
- `V1–V28`: anonymized features via PCA
- `Class`: target label — 1 = fraud, 0 = legitimate

## 2. Class Distribution

```python
# Class distribution
print(df['Class'].value_counts(normalize=True))

# Visual
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

The dataset is highly imbalanced — only ~0.17% of transactions are fraud. This will require special handling during modeling (e.g. SMOTE, class weighting).

## 3. Distribution of Amount and Time

```python
# Transaction Amount
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.show()

# Transaction Time
sns.histplot(df['Time'], bins=50, kde=True)
plt.title('Distribution of Transaction Time')
plt.show()
```

- `Amount` is right-skewed — most transactions are small.
- `Time` increases linearly and may benefit from normalization or feature engineering later.


## 4. Distribution of PCA Features (V1–V28)


```python
# Plot histograms of PCA features
df.drop(['Time', 'Amount', 'Class'], axis=1).hist(figsize=(20, 20), bins=30)
plt.tight_layout()
plt.show()
```

These features appear normally distributed, as expected from PCA. Some may be more informative for fraud detection than others.

## 5. Correlation Matrix

```python
# Correlation heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(), cmap='coolwarm', center=0, linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
# Correlation of features with the target
correlation_with_class = df.corr()['Class'].drop('Class').sort_values(ascending=False)
correlation_with_class
```

We computed the Pearson correlation between each feature and the target variable `Class` (1 = fraud, 0 = non-fraud).

Most features show low correlation, which is expected due to the PCA transformation. However, some features show moderate correlations with fraud:

### 📉 Top Negative Correlations (fraud indicators)
- **V17**: -0.33
- **V14**: -0.30
- **V12**: -0.26
- **V10**: -0.22
- **V16**: -0.20

These features may be **strong predictors of fraud**, and we will explore them visually in the next section.

### 📈 Top Positive Correlations (weaker)
- **V11**: +0.15
- **V4**: +0.13

These may still contribute to modeling, though with less predictive power.


## 6. Feature Behavior by Class

```python
important_features = ['V17', 'V14', 'V12', 'V10']

for col in important_features:
    plt.figure(figsize=(10, 4))
    sns.kdeplot(data=df, x=col, hue='Class', fill=True, common_norm=False, alpha=0.5)
    plt.title(f'Distribution of {col} by Class')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

We now visualize the distribution of key features for fraud vs. non-fraud transactions. These plots help confirm that:
- Fraudulent transactions tend to have lower values in features like V17 and V14
- These features are likely to be strong predictors in our models

## 7. Key Observations

- The dataset is extremely imbalanced (~0.17% fraud cases)
- Most features are PCA components and show little multicollinearity
- Several features (V17, V14, V12, V10) show clear separation between fraud and non-fraud cases
- `Amount` is right-skewed; `Time` shows periodic behavior
